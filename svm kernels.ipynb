```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.linspace(-5.0,5.0,100)
y=np.sqrt(10**2- X**2)
y=np.hstack([y,-y])
x=np.hstack([X,-X])
```

```python
X1=np.linspace(-5.0,5.0,100)
y1=np.sqrt(5**2- X1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([X1,-X1])
```

```python
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
df1=pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y']=0
```
